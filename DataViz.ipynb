```python
import requests
import pandas as pd 
import folium
```

```python
url = "https://lefooding.com/fr/search?type=restaurant&page=1&sort=&per_page=877&total_pages=1"
r=requests.get(url)
d=r.json()
```

```python
d['results'][0]['sorted_tags']['transport']
name = []
postal_code = []
position = []
kitchen = []
transport = []
```

```python
for i in range(len(d['results'])): 
    name.append(d['results'][i]['name'])
    postal_code.append(d['results'][i]['address']['postal_code'])
    position.append(d['results'][i]['address']['position'])
    try:
        kitchen.append(d['results'][i]['sorted_tags']['kitchen'])
    except:
        kitchen.append(" ")
    try:    
        transport.append(d['results'][i]['sorted_tags']['transport'])
    except:
        transport.append(" ")
```

```python
data = pd.DataFrame({'Postal_Code': postal_code , 'Position': position , 'Kitchen': kitchen , 'Transport':transport } , index=name)
```

```python
data 
```

```python
m = folium.Map(location=data['Position'][0],
              zoom_start=12,
              titles='Stamen Terrain')

i = 0 
while (i < len(data)):
    try:
        folium.Marker(location = data['Position'][i]).add_to(m)
        i += 1
    except:
        i += 1
```
